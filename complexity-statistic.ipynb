```python
import pandas as pd
import matplotlib.pyplot as plt

# Dữ liệu
X1 = 2.739205
X2 = 1.288835
X3 = 6.4691
X4 = 4.311655

Y1 = 49.45
Y2 = 387.6
Y3 = 90
Y4 = 75.3

data = {
    'Thuật toán': ['BFS', 'DFS', 'A*', 'A* Cải tiến'],
    'Thời gian (ms)': [X1, X2, X3, X4],
    'Độ dài (nodes)': [Y1, Y2, Y3, Y4]
}

# Tạo DataFrame từ dữ liệu
df = pd.DataFrame(data)

# In bảng dữ liệu
print("Bảng dữ liệu gốc:")
print(df)

# Tạo 2 hình riêng biệt
fig, ax = plt.subplots(figsize=(7, 5))

# Biểu đồ thời gian
ax.bar(df['Thuật toán'], df['Thời gian (ms)'], color='skyblue')
ax.set_title('Thời gian thực thi (ms)')
ax.set_xlabel('Thuật toán')
ax.set_ylabel('Thời gian (ms)')

# Thêm ghi chú vào biểu đồ thời gian
for bar in ax.patches:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom')

plt.tight_layout()
plt.show()

# Tạo hình thứ 2 cho độ dài đường đi
fig, ax = plt.subplots(figsize=(7, 5))

# Biểu đồ không gian
ax.bar(df['Thuật toán'], df['Độ dài (nodes)'], color='lightgreen')
ax.set_title('Độ dài đường đi (nodes)')
ax.set_xlabel('Thuật toán')
ax.set_ylabel('Độ dài (nodes)')

# Thêm ghi chú vào biểu đồ không gian
for bar in ax.patches:
    yval = bar.get_height()
    ax.text(bar.get_x() + bar.get_width()/2, yval, round(yval, 2), ha='center', va='bottom')

plt.tight_layout()
plt.show()
```
